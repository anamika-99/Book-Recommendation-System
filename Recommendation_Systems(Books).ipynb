## Weighted Average

```python
# Create column Rating average 
df['Avg_Rating']=df.groupby('ISBN')['Book_Rating'].transform('mean')
# Create column Rating sum
df['Users_Rated']=df.groupby('ISBN')['Book_Rating'].transform('count')
df.head(10)
```

#### Weighted Rating(WR)=[vR/(v+m)]+[mC/(v+m)] where, v is the number of ratings for the books m is the minimum ratings required to be listed in the chart R is the average rating of the book C is the mean rating across the whole report.

```python
C= df['Avg_Rating'].mean()
m= df['Users_Rated'].quantile(0.90)
v = df['Users_Rated']
R = df['Avg_Rating']
C,m # minimum 64 users should have rated the book
```

```python
#creating dataset with m condition.
data_m=df[df['Users_Rated'] >= m]
data_m
```

```python
data_m['Score']=(v/(v+m) * R) + (m/(m+v) * C)
data_m
```

## Knowledge-based recommendation function

```python
def knowledge_based_recommendation(user_profile, df):
    # Filter books based on user profile
    filtered_books = df.copy()
    if 'age' in user_profile:
        filtered_books = filtered_books[filtered_books['Age'] >= user_profile['age']]
    if 'country' in user_profile:
        filtered_books = filtered_books[filtered_books['Country'] == user_profile['country']]
    
    # Sort books based on some criteria (e.g., average rating)
    recommended_books = filtered_books.sort_values(by='Avg_Rating', ascending=False).head(5)
    
    return recommended_books[['ISBN', 'Book-Title', 'Book-Author', 'Avg_Rating', 'Users_Rated']]
```

```python
# Example usage
user_profile = {'age': 32, 'country': 'italy'}  # Example user profile
recommended_books = knowledge_based_recommendation(user_profile, df)
print(recommended_books)
```

## Demographic-based recommendation function 

```python
def demographic_based_recommendation(age, country, df):
    # Filter books based on demographic information
    filtered_books = df[(df['Age'] == age) & (df['Country'] == country)]
    
    # Sort books based on some criteria (e.g., average rating)
    recommended_books = filtered_books.sort_values(by='Avg_Rating', ascending=False).head(5)
    
    return recommended_books[['ISBN', 'Book-Title', 'Book-Author', 'Avg_Rating', 'Users_Rated']]
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(top_ratings, test_size=0.20)
```

```python
train_data
```

```python
u_unique_train = train_data['User-ID'].unique()  
train_data_user2idx = {x:i for i, x in enumerate(u_unique_train)}

# Get int mapping for isbn in train dataset
i_unique_train = train_data.ISBN.unique()  
train_data_book2idx = {y:i for i, y in enumerate(i_unique_train)}
```

```python
# Get int mapping for user_id in test dataset
u_unique_test = test_data['User-ID'].unique()  
test_data_user2idx = {x:i for i, x in enumerate(u_unique_test)}

# Get int mapping for isbn in test dataset
i_unique_test = test_data.ISBN.unique() 
test_data_book2idx = {y:i for i, y in enumerate(i_unique_test)}
```

```python
train_data['u_unique'] = train_data['User-ID'].map(train_data_user2idx)
train_data['i_unique'] = train_data['ISBN'].map(train_data_book2idx)

# testing set
test_data['u_unique'] = test_data['User-ID'].map(test_data_user2idx)
test_data['i_unique'] = test_data['ISBN'].map(test_data_book2idx)

# Convert back to three feature of dataframe 
train_data = train_data[['u_unique', 'i_unique', 'Book_Rating']]
test_data = test_data[['u_unique', 'i_unique', 'Book_Rating']]
```

```python
train_data
```

```python
# first create an empty matrix of users books and then add the appropriate values to the matrix by extracting them from the dataset
n_users = train_data['u_unique'].nunique()
n_books = train_data['i_unique'].nunique()

train_matrix = np.zeros((n_users, n_books))

for entry in train_data.itertuples():                  
    train_matrix[entry[1]-1, entry[2]-1] = entry[3]    # entry[1] is the user-id, entry[2] is the book-isbn and -1 is to counter 0-based indexing
```

```python
train_matrix.shape
```

```python
n_users = test_data['u_unique'].nunique()
n_books = test_data['i_unique'].nunique()

test_matrix = np.zeros((n_users, n_books))

for entry in test_data.itertuples():
    test_matrix[entry[1]-1, entry[2]-1] = entry[3] 
```

```python
test_matrix
```

```python
# To make item-item similarity we need to take the transpose of the matrix

from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(train_matrix, metric='cosine')
item_similarity = pairwise_distances(train_matrix.T, metric='cosine')
```

```python
# function to predict the similarity :
def predict_books(ratings, similarity, type='user'): # default type is 'user'
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        
        # Use np.newaxis so that mean_user_rating has the same format as ratings
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
item_prediction = predict_books(train_matrix, item_similarity, type='item')
user_prediction = predict_books(train_matrix, user_similarity, type='user')
```

# Evaluation

```python
# Evaluation metric by mean squared error
from sklearn.metrics import mean_squared_error
from math import sqrt

def rmse(prediction, test_matrix):
    prediction = prediction[test_matrix.nonzero()].flatten()
    test_matrix = test_matrix[test_matrix.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, test_matrix))

print(f'Item-based CF RMSE: {rmse(item_prediction, test_matrix)}')
print(f'User-based CF RMSE: {rmse(user_prediction, test_matrix)}')
```
